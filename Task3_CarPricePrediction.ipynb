```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
df = pd.read_csv('CarPrice.csv')
```

```python
from sklearn.preprocessing import OneHotEncoder
numerical_data=df[["symboling", "wheelbase", "carlength","carwidth", "carheight", "curbweight","enginesize", "boreratio", "stroke","compressionratio", "horsepower", "peakrpm","citympg", "highwaympg"]]
categorical_data = df[["CarName","fueltype","aspiration","doornumber","carbody","drivewheel","enginelocation","enginetype","cylindernumber","fuelsystem"]]
encoder = OneHotEncoder()
one_hot_encoded_data = encoder.fit_transform(categorical_data)
one_hot_encoded_df = pd.DataFrame(one_hot_encoded_data.toarray(), columns=encoder.get_feature_names_out(input_features=categorical_data.columns))
data = df.drop(columns=["CarName","fueltype","aspiration","doornumber","carbody","drivewheel","enginelocation","enginetype","cylindernumber","fuelsystem"])
data = pd.concat([data, one_hot_encoded_df], axis=1)
```

```python
data_final=data.drop(columns=["car_ID","price"])
x = np.array(data_final)
y = np.array(data['price'])
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
model = DecisionTreeRegressor()
model.fit(xtrain, ytrain)
predictions = model.predict(xtest)
```

```python
train_accuracy=model.score(xtrain,ytrain)
test_accuracy= model.score(xtest,ytest)
print("The training accuracy is", train_accuracy)
print("The testing accuracy is", test_accuracy)
```
